# 01 - Criação e Manipulação de Arrays

Crie um array Numpy com 10 números inteiros aleatórios entre 1 e 20 e extraia os elementos pares.

```python
import numpy as np

array = np.random.randint(1, 20, size = 10)

pares = array[array % 2 == 0]

print("Pares: ", pares)
```

# 06 - Criação de DataFrames

Crie um DataFrame com dados de 5 pessoas, onde cada pessoa tem nome, idade (entre 20 e 40), e cidade (escolha 3 cidades aleatórias).

```python
import pandas as pd
import random

nomes = ["João", "Maria", "Pedro", "Ana", "Lucas"]
idades = [random.randint(20, 40) for _ in range(5)]
cidades = ["São Paulo", "Rio de Janeiro", "Belo Horizonte"]

dados = {
    "Nome": nomes,
    "Idade": idades,
    "Cidade": [random.choice(cidades) for _ in range(5)]
}

df = pd.DataFrame(dados)

print(df)
```

# 09 - Operações em Colunas

Adicione uma coluna ao DataFrame com o valor total da venda (Quantidade * Preço).

```python
import pandas as pd
import numpy as np

num_linhas = 10

produtos = [f'Produto_{i+1}' for i in range(num_linhas)]

quantidade = np.random.randint(1, 21, size=num_linhas)
preco = np.random.uniform(10, 101, size=num_linhas).round(2)

df_vendas = pd.DataFrame({
    'Produto': produtos,
    'Quantidade': quantidade,
    'Preço': preco
})

df_vendas['Valor Total'] = df_vendas['Quantidade'] * df_vendas['Preço']

print("DataFrame com Valor Total:")
print(df_vendas)
```

# 10 - Agrupamento e Agregação

Agrupe o DataFrame pelo campo Produto e calcule a soma das quantidades vendidas para cada produto.

```python
import pandas as pd
import numpy as np

num_linhas = 10

produtos = [f'Produto_{i+1}' for i in range(num_linhas)]

quantidade = np.random.randint(1, 21, size=num_linhas)
preco = np.random.uniform(10, 101, size=num_linhas).round(2)

df_vendas = pd.DataFrame({
    'Produto': produtos,
    'Quantidade': quantidade,
    'Preço': preco
})

df_vendas['Valor Total'] = df_vendas['Quantidade'] * df_vendas['Preço']

df_agrupado = df_vendas.groupby('Produto')['Quantidade'].sum().reset_index()

print("DataFrame Agrupado por Produto com Soma das Quantidades:")
print(df_agrupado)
```

# 11 - Gráfico de Linhas

Crie um gráfico de linha que mostra a variação de preços de um produto fictício ao longo de 12 meses.

```python
import numpy as np
import matplotlib.pyplot as plt

meses = np.arange(1, 13)
precos = np.random.uniform(10, 100, size=12).round(2)

plt.figure(figsize=(10, 6))
plt.plot(meses, precos, marker='o', linestyle='-', color='g', label='Preço do Produto')

plt.title('Variação de Preços do Produto ao Longo de 12 Meses')
plt.xlabel('Meses')
plt.ylabel('Preço (R$)')

plt.grid(True)

plt.legend()

plt.show()
```

# 12 - Gráfico de Barras

# 14 - Gráfico de Dispersão

Crie um gráfico de dispersão que mostre a relação entre altura e peso em um conjunto de dados fictício.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

altura = np.random.normal(170, 10, 100)
peso = np.random.normal(70, 5, 100)

plt.figure(figsize=(10, 6))
plt.scatter(altura, peso, color='green', alpha=0.4)

plt.title('Relação entre Altura e Peso')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')

plt.grid(True)

plt.show()
```

# 15 - Customização de Gráficos

Adicione título, rótulos de eixo e uma legenda ao gráfico de barras criado anteriormente.

```python
import matplotlib.pyplot as plt

produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [25, 40, 15, 30, 50]

plt.figure(figsize=(10, 6))
bars = plt.bar(produtos, quantidades, color='lightgreen', label='Quantidade Vendida')

plt.title('Quantidade Vendida de Produtos Fictícios')
plt.xlabel('Produtos')
plt.ylabel('Quantidade Vendida')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 1, str(height), ha='center', va='bottom')

plt.legend()

plt.show()
```

# 16 - Gráfico de Barras com Seaborn

Use Seaborn para criar um gráfico de barras com os mesmos dados fictícios de quantidade vendida por produto.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [25, 40, 15, 30, 50]

df_vendas = pd.DataFrame({
    'Produto': produtos,
    'Quantidade': quantidades
})

plt.figure(figsize=(10, 6))
sns.barplot(x='Produto', y='Quantidade', data=df_vendas, palette='viridis')

plt.title('Quantidade Vendida de Produtos Fictícios')
plt.xlabel('Produtos')
plt.ylabel('Quantidade Vendida')

plt.show()
```

# 17 - Gráfico de Dispersão com Seaborn

Recrie o gráfico de dispersão da relação entre altura e peso usando Seaborn, incluindo uma linha de regressão.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

np.random.seed(42)
altura = np.random.normal(170, 10, 100)
peso = np.random.normal(70, 5, 100)

df = pd.DataFrame({
    'Altura': altura,
    'Peso': peso
})

plt.figure(figsize=(10, 6))
sns.regplot(x='Altura', y='Peso', data=df, scatter_kws={'color':'green', 'alpha':0.5}, line_kws={'color':'red'})

plt.title('Relação entre Altura e Peso com Linha de Regressão')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')

plt.show()
```

# 19 - Heatmap

Crie um heatmap para mostrar a correlação entre diferentes variáveis em um DataFrame fictício.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

num_amostras = 100
num_variaveis = 5

np.random.seed(42)
dados = np.random.rand(num_amostras, num_variaveis)

df = pd.DataFrame(dados, columns=[f'Variável_{i+1}' for i in range(num_variaveis)])

correlacao = df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlacao, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, vmin=-1, vmax=1)

plt.title('Heatmap da Correlação entre Variáveis')

plt.show()
```

# 20 - Pairplot

Crie um pairplot para explorar relações entre várias variáveis em um conjunto de dados fictício.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

num_amostras = 150
num_variaveis = 4

np.random.seed(42)
dados = np.random.rand(num_amostras, num_variaveis) * 100

df = pd.DataFrame(dados, columns=[f'Variável_{i+1}' for i in range(num_variaveis)])

pairplot = sns.pairplot(df, diag_kind='kde', markers='o', palette='viridis')

plt.suptitle('Pairplot das Variáveis', y=1.02)

plt.show()
```
